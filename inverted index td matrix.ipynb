```python
import PyPDF2 
doc1 = open(r'C:\Users\AMRITHA R J\Downloads\omicron.pdf', 'rb') 
doc2 = open(r'C:\Users\AMRITHA R J\Downloads\t20.pdf', 'rb') 
doc3 = open(r'C:\Users\AMRITHA R J\Downloads\What is a Heart Attack.pdf', 'rb') 
```

```python
pdf1 = PyPDF2.PdfFileReader(doc1)
pdf2 = PyPDF2.PdfFileReader(doc2) 
pdf3 = PyPDF2.PdfFileReader(doc3) 
```

```python
import nltk
nltk.download('punkt')
```

```python
m = pdf1.numPages
for i in range(m):
    page1 = pdf1.getPage(i)
    p = page1.extractText()
    
n = pdf2.numPages
for i in range(n):
    page2 = pdf2.getPage(i)
    q = page2.extractText()

o = pdf3.numPages
for i in range(o):
    page3 = pdf3.getPage(i)
    r = page3.extractText()
    
print(p)
print(q)
print(r)
```

```python
docs = [p,q,r]
docs
```

```python
unique_terms = {term for doc in docs for term in doc.split()}
unique_terms
```

```python
#Term document matrix
```

```python
doc_term_matrix = {}

for term in unique_terms:
    doc_term_matrix[term] = []
    
    for doc in docs:
        if term in doc:
            doc_term_matrix[term].append(1)
        else: doc_term_matrix[term].append(0)

doc_term_matrix
```

```python
import numpy as np
docs_array = np.array(docs, dtype='object')

v1 = np.array(doc_term_matrix['COVID'])    
v2 = np.array(doc_term_matrix['T20'])
v3 = np.array(doc_term_matrix['heart'])
print(v1)
print(v2)
print(v3)
```

```python
v = v1 & v2 & v3
print(v)
```

```python
v1 = np.array(doc_term_matrix['age'])    
v2 = np.array(doc_term_matrix['nationalities'])

print(v1)
print(v2)
```

```python
v = v1 | v2
print(v)
```
